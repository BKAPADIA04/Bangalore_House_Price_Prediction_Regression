```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/final_df.csv')
df.head()
```

```python
X = df.drop(columns = ['price'])
```

```python
y = df['price']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

**Ordinary Least Square Approach**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
y_pred = lr.predict(X_test)
print("R2Score = ",r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge,Lasso,ElasticNet
```

**Linear Regression**

```python
R_0 = Ridge(alpha=0)
R_0.fit(X_train,y_train)
print("R2Score = ",r2_score(y_test,R_0.predict(X_test)))
```

**For Ridge Regression, the maximum R2 score was achieved at alpha = 0.1**

```python
R_10 = Ridge(alpha=0.1)
R_10.fit(X_train,y_train)
print("R2Score = ",r2_score(y_test,R_10.predict(X_test)))
```

**Lasso Regression**

```python
L_0 = Lasso(alpha=0)
L_0.fit(X_train,y_train)
print("R2Score = ",r2_score(y_test,L_0.predict(X_test)))
```

**For Lasso Regression, the maximum R2 score was achieved at alpha = 0.001**

```python
L_10 = Lasso(alpha = 0.001)
L_10.fit(X_train,y_train)
print("R2Score = ",r2_score(y_test,L_10.predict(X_test)))
```

**For ElasticNet Regression, the maximum R2 score was achieved at alpha = 0.0005 and l1_ratio = 0.8**

```python
reg = ElasticNet(alpha=0.0005,l1_ratio=0.8)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

**The R2 score ranges from 0.853 to 0.854 in Linear,Ridge,Lasso and ElasticNet Regression.**

***So we will select ***Linear Regression*** as our model.***

```python
def predict_price(location,sqft,bhk,bath):
    index_location = np.where(X.columns == location)[0][0]
    
    features = np.zeros(len(X.columns))
    features[0] = sqft
    features[1] = bath
    features[2] = bhk
    
    if index_location >= 1 :
        features[index_location] = 1
    
    return lr.predict([features])[0]
```

```python
predict_price('Electronic City',1000,2,2)
```
