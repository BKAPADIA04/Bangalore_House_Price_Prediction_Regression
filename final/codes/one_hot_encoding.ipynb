**Applying One Hot Encoding on Location Column.**

```python
pd.get_dummies(df_copy1,columns=['location'])
```

```python
pd.get_dummies(df_copy1.location,dtype='int')
```

```python
ohe = pd.get_dummies(df_copy1.location,dtype='int')
ohe
```

```python
ohe.drop(columns=['other'],inplace=True)
```

```python
df_copy1.shape
```

```python
concat_dummy = pd.concat([df_copy1, ohe], axis=1)
concat_dummy
```

```python
(concat_dummy.location == 'other').sum()
```

```python
df = concat_dummy
```
